```python
import cobra
#Source of GEM model AGORA2
Heinken, A., Hertel, J., Acharya, G. et al. Genome-scale metabolic reconstruction of 7,302 human microorganisms for personalized medicine. Nat Biotechnol 41, 1320â€“1331 (2023). https://doi.org/10.1038/s41587-022-01628-0

model_recta= cobra.io.read_sbml_model("/Users/amin/Documents/Research/CrossFeedingPaper2/GEMS/Eubacterium_rectale_ATCC_33656.xml")
shwa= cobra.io.read_sbml_model("/Users/amin/Documents/Research/Codes/ABM_code/gutagentbased/GEM_ABM/Shewanella_oneidensis_MR-1_FinalGapfill.xml")
bacilus=cobra.io.read_sbml_model("/Users/amin/Documents/Research/Codes/ABM_code/gutagentbased/PostProcessing/Shewanella_oneidensis_MR-1_GP1.xml")
from cobra.io import write_sbml_model
```

```python
def exchange_reactions(model):
    exchange_reactions = [reaction for reaction in model_recta.reactions if reaction.id.startswith('EX_')]
    return exchange_reactions

#Extracting and printing all carbon compounds in the model
def carbon_compounds(model):
    carbon_compounds = []
    # Iterate through metabolites to find those containing carbon
    for metabolite in model.metabolites:
        if 'C' in metabolite.elements:  # Check if carbon is present in the metabolite's elemental composition
            carbon_compounds.append(metabolite)
    for metabolite in carbon_compounds:
        print(f"{metabolite.id}: {metabolite.name} - Formula: {metabolite.formula}")
        print('number of carbon sources', len(carbon_compounds))
    return carbon_compounds

#Extracting and printing all corbon sources in the media in the model
def carbon_sources(model): 
    carbon_sources=[]
    exchange_rxns=exchange_reactions(model)
    for rxn in exchange_rxns:
        for metabolite in rxn.metabolites:
            if 'C' in metabolite.elements:
                carbon_sources.append(metabolite)
        for metabolite in carbon_sources:
            print(f"{metabolite.id}: {metabolite.name} - Formula: {metabolite.formula}")
            print('number of carbon sources', len(carbon_sources))
    return carbon_sources

#adjusting the rection bounds for carbon sources
def reactionbound(model, carbonupper, carbonlower):
    exchange_rxns=exchange_reactions(model)
    carbon_sources1=carbon_sources(model)
    #print(carbon_souorces)
    for rxn1 in exchange_rxns:
        for met in rxn1.metabolites:
            if met in carbon_sources1:
                exrxnid=model.reactions.get_by_id(rxn1.id)
                print(rxn1.lower_bound)
                print(rxn1.upper_bound)
                exrxnid.lower_bound=carbonlower
                exrxnid.upper_bound=carbonupper
                if "glucose" in met.name:
                    print('find me')
                    print(met)
                    print(rxn1.id)
                    exrxnid.lower_bound=carbonlower
                    exrxnid.upper_bound=carbonupper

def print_biomass(model):

    biomass_rxn = [rxn for rxn in model.reactions if 'biomass' in rxn.id.lower() or 'biomass' in rxn.name.lower()]
    print('number of biomass reactions', len(biomass_rxn))
    j=0
    for rxn in biomass_rxn:
        print(j)
        print('reaction id:', rxn.id, 'LL', '\n')
        j=j+1
        
        print('reaction id:', 'valiei', 'LL', '\n')
        #print('reaction  name', biomass_rxn[0].name)
        print(rxn.metabolites)
        print('reaction name', rxn.name)
        print('rxn id', rxn.id, '\n')
    print_biomass(model_recta)
    
```
